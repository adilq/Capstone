# waypoints

```python
import numpy as np
 
waypoints = np.array([[-5.24511607e-03, -1.82299304e-03,  5.00000000e-01],
                      [ 1.80000000e+00,  1.90000000e+00,  5.00000000e-01],
                      [ 1.80000000e+00, -1.80000000e+00,  5.00000000e-01],
                      [-1.80000000e+00, -1.60000000e+00,  5.00000000e-01],
                      [-1.70000000e+00,  2.00000000e+00,  5.00000000e-01]])
```

```python
waypoints
```

# incremental

```python
NAV_THRESHOLD = 0.5
NAV_INCREMENT = 0.5

test_waypoints = waypoints[0].reshape(1, 3)

for i, pos in enumerate(waypoints[1:]):
    distance = pos - waypoints[i-1]
    length = np.linalg.norm(distance)
    
    if length > NAV_THRESHOLD:
        unit = distance/length 
        
        num_steps = round(length / NAV_INCREMENT)
        steps = np.linspace(waypoints[i], pos, num_steps+1)
        
        test_waypoints = np.vstack([test_waypoints, steps[1:]])
        
```
